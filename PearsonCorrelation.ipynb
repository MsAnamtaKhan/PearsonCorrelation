```python
import math
def pearsonCorrelation(x,y):
    xStd=np.std(x)
    yStd=np.std(y)
    cov=np.cov(x,y)
    r=cov[0][1]/((xStd)*(yStd))
    return np.round(r,2)
  
```

```python
n=len(col)
n
```

```python
matrix=[]

for i in range(n):          
    a=[]
    for j in range(n):      
         a.append(pearsonCorrelation(X[col[i]],X[col[j]]))
    matrix.append(a)
```

```python
a=[]
for i in range(n):
    a.append(col[i])

import seaborn as sns
plt.figure(figsize=[15,15])
plt.title('Pearson Correlation Matrix', fontsize = 20)
sns.heatmap(matrix,annot=True,xticklabels=a,yticklabels=a)
```
